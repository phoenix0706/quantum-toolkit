## Double-Slit Interference and Decoherence (Computational Demo)

This notebook demonstrates one of the most famous experiments in physics — the **Double-Slit Experiment** — and explores how **quantum interference arises from wave superposition**.

Using a simple computational model, we will study:

- Classical vs Quantum probability
- Wave interference
- Quantum superposition
- Decoherence (loss of interference)

These ideas form the foundation of:

- Quantum Mechanics

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=(8,5)
plt.rcParams["font.size"]=12

print("Notebook Initialized")
````

# 1. Historical Background

The **Double-Slit Experiment** was first performed by **Thomas Young (1801)** to demonstrate the wave nature of light.

Later, the experiment was repeated using:

- electrons
- neutrons
- atoms
- even large molecules

Surprisingly, even when particles are fired **one at a time**, an **interference pattern still appears** over time.

This suggests that each particle behaves like a **probability wave**.

This phenomenon is one of the clearest demonstrations of the strange nature of **quantum mechanics**.

The double-slit experiment demonstrates that particles such as electrons, photons, and atoms behave like waves.

When particles pass through two slits:

- Classical physics prediction: particles choose one slit → probabilities add.-

- Quantum physics prediction: particles exist in a superposition of both paths → amplitudes add.

This leads to an interference pattern on the screen.


### Physical Parameters

```
wavelength = 1.0
k = 2π/λ
d = slit separation
L = distance to screen
```

| Parameter | Meaning                       |
| --------- | ----------------------------- |
| λ         | wavelength of particle        |
| k         | wave number                   |
| d         | distance between slits        |
| L         | distance from slits to screen |

````python
#screen coordinate
x=np.linspace(-10,10,2000)

#Physical Parameters
wavelength=1.0
k=2*np.pi/wavelength 
d=2.0 # slit separation
L=10 # distance to screen

#Complex amplitudes from two slits
psi1=np.exp(1j*k*np.sqrt((x-d/2)**2+L**2))
psi2=np.exp(1j*k*np.sqrt((x+d/2)**2+L**2))

#Quantum interference
P_quantum=np.abs(psi1+psi2)**2

#Classical probability addition
P_classical=np.abs(psi1)**2+np.abs(psi2)**2

plt.plot(x, P_quantum, label="Quantum (|ψ1+ψ2|²)")
plt.plot(x,P_classical,"--",label="Classical (|ψ1|² + |ψ2|²)")
plt.xlabel("Screen Position")
plt.ylabel("Probability (unnormalized)")
plt.title("Double-Slit: Classical vs Quantum")
plt.legend()
plt.show()
````

### Explicit Interference Term

````python
cross_term=2*np.real(np.conj(psi1)*psi2)
plt.plot(x, cross_term)
plt.title("Interference Cross Term: 2 Re(ψ1* ψ2)")
plt.xlabel("Screen Position")
plt.ylabel("Cross-term contribution")
plt.show()

print("Notice: Interference comes entirely from the cross term")
````

## Decoherence (Which-Path Information)
DECOHERENCE PARAMETER γ

 γ = 1 → full coherence (full interference)
 
 γ = 0 → full which-path information -> no interference

 γ = 0.5 → partial interference


````python
gamma=0
P_decoherence=(np.abs(psi1)**2+
               np.abs(psi2)**2+
               gamma*2*np.real(np.conj(psi1)*psi2)
              )

plt.plot(x, P_decoherence)
plt.title(f"Decoherence with γ = {gamma}")
plt.xlabel("Screen Position")
plt.ylabel("Probability")
plt.show()

print("γ reduces the interference term.")
````

````python
gamma=1 # no noise - ideal system
P_decoherence=(np.abs(psi1)**2+
               np.abs(psi2)**2+
               gamma*2*np.real(np.conj(psi1)*psi2)
              )

plt.plot(x, P_decoherence)
plt.title(f"Decoherence with γ = {gamma}")
plt.xlabel("Screen Position")
plt.ylabel("Probability")
plt.show()


````

````python
gamma=0.3
P_decoherence=(np.abs(psi1)**2+
               np.abs(psi2)**2+
               gamma*2*np.real(np.conj(psi1)*psi2)
              )

plt.plot(x, P_decoherence)
plt.title(f"Decoherence with γ = {gamma}")
plt.xlabel("Screen Position")
plt.ylabel("Probability")
plt.show()


````

````python
gamma=0.5
P_decoherence=(np.abs(psi1)**2+
               np.abs(psi2)**2+
               gamma*2*np.real(np.conj(psi1)*psi2)
              )

plt.plot(x, P_decoherence)
plt.title(f"Decoherence with γ = {gamma}")
plt.xlabel("Screen Position")
plt.ylabel("Probability")
plt.show()


````
