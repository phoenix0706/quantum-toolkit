## STERN GERLACH IN 2D Hilbert Space

```python
import numpy as np
```

```python
# Define Z basis
up_z = np.array([1, 0], dtype=complex) #spin up along Z
down_z = np.array([0, 1], dtype=complex) # spin down along Z
# these are measurement outcomes- measurement along z basis)
 
# Define X basis (measurement along X axis)
# superposition states
up_x= (1/np.sqrt(2))*np.array([1,1], dtype=complex)
down_x= (1/np.sqrt(2))*np.array([1,-1], dtype=complex)
```

```python
up_z, down_z, up_x, down_x
```

```python
# implementation of born rule
def probability(state, outcome):
    return np.abs(np.vdot(outcome, state))**2

print("Hilbert space initialized.")
```

```python
# First Stern–Gerlach Measurement (P=∣⟨ϕ∣ψ⟩∣2)
# MEASURE |-z> IN X BASIS

print("P(+x | +z) =", probability(down_z, up_x))
print("P(-x | +z) =", probability(down_z, down_x))
```

```python
collapsed_state=up_x #measuring in x basis, state collapse
print("P(+z | +x)=", probability(collapsed_state, up_z))
print("P(-z | +x)=", probability(collapsed_state, down_z))
```

```python
# Pauli Matrices & Non commutation
sigma_x=np.array([[0,1], [1,0]], dtype=complex)
sigma_z=np.array([[1,0],[0,-1]], dtype=complex)
commutator=sigma_x @ sigma_z - sigma_z @ sigma_x
print('Commutator [σx, σz]=')
print(commutator)
```

In quantum mechanics:

If two operators commute → you can measure both precisely
If they don’t commute → measuring one disturbs the other

So here:

- Measuring X-spin destroys information about Z-spin
- Measuring Z-spin destroys information about X-spin

This is exactly what we observed earlier in our code:

Z → X → Z gives randomness

```python
# unification (Density Matrix view)

#pure superposition state
psi=(1/np.sqrt(2))*np.array([1,1],dtype=complex)
#density matrix
rho=np.outer(psi, np.conj(psi))
print("Pure state density matrix:")
print(rho)

# decohered version(remove off diagonal terms)
rho_deco=np.array([[rho[0,0],0], [0,rho[1,1]]], dtype=complex)

print("\nDecohered density matrix")
print(rho_deco)

# Off-diagonal terms = coherence
# Remove them → no interference
# Same physics as double-slit γ = 0
```

- Quantum information is stored in off-diagonal terms
- Measurement/environment → destroys coherence
- That’s why quantum → classical transition happens
